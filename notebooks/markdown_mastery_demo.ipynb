# 4. Effective Markdown-Code Cell Alternation

**The golden rule**: Use Markdown to explain *why*, use code to show *how*.

## Pattern: Explain → Execute → Interpret

### Step 1: Explain Intent (Markdown)

We'll calculate basic statistics for our dataset to understand the central tendency and spread of the scores.

````python
# Step 2: Execute the analysis
mean_score = df['score'].mean()
median_score = df['score'].median()
std_score = df['score'].std()

print(f"Mean score: {mean_score:.1f}")
print(f"Median score: {median_score:.1f}")
print(f"Standard deviation: {std_score:.1f}")
````

### Step 3: Interpret Results (Markdown)

The statistical analysis reveals:

- **Mean (85.8)**: The average score is quite high
- **Median (86.5)**: Close to mean, suggesting symmetric distribution
- **Std Dev (5.1)**: Low variability indicates consistent performance

This suggests the group performs consistently well with minimal outliers.

## Common Anti-Patterns to Avoid

❌ **Don't**: Put explanations in code comments

```python
# This is bad - should be in Markdown
x = 5  # Initialize x with value 5
y = 10 # Initialize y with value 10
result = x + y # Add x and y together
```

✅ **Do**: Use Markdown for explanations

We initialize two variables and calculate their sum to demonstrate basic arithmetic operations.

````python
# Clean, executable code without excessive comments
x = 5
y = 10
result = x + y
print(f"Result: {result}")
````
