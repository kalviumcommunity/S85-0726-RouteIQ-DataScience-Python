# Kernel Control Demo Notebook

This notebook demonstrates safe and intentional Jupyter kernel control: normal execution, interrupting a long-running cell, restarting the kernel, and rerunning from top.

## Step 1: Execute cells in order and observe retained state

The first cell defines a variable. The next cell uses it, proving the kernel retains state across cell execution.

```python
base_value = 10
print("base_value set to:", base_value)
```

```python
print("base_value + 5 =", base_value + 5)
```

## Step 2: Start a long-running cell and interrupt safely

Run the next cell and interrupt after a few ticks using **Kernel -> Interrupt**. This demonstrates stopping execution without closing the notebook.

```python
import time
for i in range(1, 31):
    print(f"tick {i}")
    time.sleep(1)
```

## Step 3: Restart kernel, then rerun from top

After restarting the kernel, run all cells from the top to rebuild state. The next check cell verifies whether `base_value` exists in current kernel memory.

```python
print("base_value present in kernel state:", "base_value" in globals())
```
