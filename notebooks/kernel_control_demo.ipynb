# Kernel Control Demo Notebook

This notebook demonstrates **safe and intentional Jupyter kernel control**. Understanding kernel management is essential for:

- **Normal execution**: Running cells in sequence with state retention
- **Interrupting**: Safely stopping long-running operations
- **Restarting**: Clearing kernel state when needed
- **Rebuilding**: Re-executing from top after restart

**Why this matters**: Proper kernel control prevents crashes, saves time, and ensures reproducible results.

## Step 1: Execute Cells in Order and Observe Retained State

### Understanding Kernel State

The Jupyter kernel maintains **state across cell executions**. This means:

- Variables defined in one cell are available in subsequent cells
- Functions and imports persist throughout the session
- Data structures remain in memory until explicitly cleared or kernel restart

### What We'll Demonstrate

1. **Cell 2**: Define `base_value` variable
2. **Cell 3**: Access and use `base_value` to prove state retention
3. **Later**: Show how restart affects this state

This pattern mirrors real data analysis workflows where you build up state gradually.

```python
base_value = 10
print("base_value set to:", base_value)
```

```python
print("base_value + 5 =", base_value + 5)
```

## Step 2: Start a Long-Running Cell and Interrupt Safely

### Understanding Kernel Interruption

Sometimes cells run longer than expected (infinite loops, large computations, network timeouts). **Safe interruption** allows you to:

- Stop execution without closing the notebook
- Preserve existing kernel state
- Continue working after interruption

### Demonstration Steps

1. Run the next cell (30-second countdown)
2. After 3-5 seconds, use **Kernel → Interrupt** or press **I, I**
3. Observe that execution stops gracefully
4. Verify that `base_value` is still available

**Note**: This is different from restarting the kernel, which clears all state.

```python
import time
for i in range(1, 31):
    print(f"tick {i}")
    time.sleep(1)
```

## Step 3: Restart Kernel and Rebuild State

### Understanding Kernel Restart

**Kernel restart** completely clears:
- All variables from memory
- Imported libraries
- Function definitions
- Any accumulated state

### When to Restart

- **Memory issues**: Clear large datasets from memory
- **Conflicting imports**: Reload changed modules
- **Clean environment**: Start fresh for new analysis
- **Troubleshooting**: Resolve kernel corruption issues

### Rebuilding State

After restart, you must:
1. **Re-run imports**: Reload required libraries
2. **Re-execute cells**: Rebuild variables and functions
3. **Verify state**: Confirm everything is properly initialized

The next cell checks if `base_value` exists in the current kernel state.

```python
print("base_value present in kernel state:", "base_value" in globals())
```
